Las visualizaciones generadas a continuación reflejarán solo los datos disponibles.


```python
# Histograma del precio de los autos
plt.figure(figsize=(8, 4))
sns.histplot(df_spain[' Price '], bins=30, kde=True)
plt.title('Distribución del Precio de Autos en España')
plt.xlabel('Precio')
plt.ylabel('Frecuencia')
plt.grid()
plt.show()
```

El histograma muestra que hay una alta frecuencia de autos con precios cercanos a cero, lo que indica que el mercado está saturado de autos de bajo precio. A partir de los 100,000 euros, la frecuencia de venta es casi nula, sugiriendo que hay muy pocos autos en el rango de precios altos (más de 100,000 euros)

```python
# Gráfico de caja del precio por tipo de combustible
plt.figure(figsize=(10, 6))
sns.boxplot(x='ID_Fuel_Type', y=' Price ', data=df_spain)
plt.title('Precio de Autos según Tipo de Combustible')
plt.xlabel('Tipo de Combustible')
plt.ylabel('Precio')
plt.grid()
plt.show()
```

El gráfico de caja revela que el tipo de combustible 1 tiene una mayor concentración de precios bajos, mientras que el tipo 2 (siendo por diferencia el tipo de combustible mas vendido, por lo que vemos una amplia superioridad en la cantidad de datos) muestra una amplia dispersión, incluyendo autos con precios muy altos. Los tipos 3 y 4 presentan patrones similares, pero con diferentes niveles de dispersión. Esto sugiere que ciertos tipos de combustible están asociados con precios más altos, mientras que otros son más accesibles.

```python
# Gráfico de dispersión entre año, precio y kilometraje
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Year', y=' Price ', size='Kilometres', sizes=(20, 200), data=df_spain, alpha=0.5)
plt.title('Relación entre Año, Precio y Kilometraje')
plt.xlabel('Año')
plt.ylabel('Precio')
plt.grid()
plt.show()
```

El gráfico de dispersión ilustra la relación entre el año del auto, su precio y el kilometraje. A medida que se incrementa el año, se observa una tendencia general hacia precios más altos, especialmente para vehículos con bajo kilometraje. Sin embargo, hay puntos de datos que se dispersan (vemos por ejemplo un pico bastante marcado en el año 1990), lo que indica que algunos autos más viejos también tienen precios competitivos, posiblemente por su marca o condición. La variación en el tamaño de los puntos representa el kilometraje, sugiriendo que los autos con menor kilometraje tienden a tener precios más altos, mientras que los de mayor kilometraje abarcan un rango más amplio de precios

Los 3 graficos anteriores fueron creados haciendo uso de la libreria Seaborn, esto se ve reflejado en el codigo de creacion de los graficos ya que comienzan con 'sns.'

A continuacion 3 graficos haciendo uso de la libreria Matplotlib

```python
# Importar la libreria matplotlib
import matplotlib.pyplot as plt

# Calcular el precio promedio por tipo de combustible
average_price_by_fuel = df_spain.groupby('ID_Fuel_Type')[' Price '].mean()

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
average_price_by_fuel.plot(kind='bar', color='skyblue')
plt.title('Precio Promedio de Autos por Tipo de Combustible')
plt.xlabel('Tipo de Combustible')
plt.ylabel('Precio Promedio')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

#Este grafico me permite comparar el precio promedio de los autos según su tipo de combustible en España.
```

El gráfico de barras muestra claramente la variación en el precio promedio de los autos según el tipo de combustible. Se observa que los tipos de combustible como el 3 y el 4 tienen precios significativamente más altos, lo que sugiere que estos tipos de autos son más costosos en el mercado español. Por otro lado, los combustibles 1, 5 y 7 presentan precios más bajos, indicando posiblemente una menor demanda o una oferta más amplia de estos vehículos.

```python
# Histograma del kilometraje
plt.figure(figsize=(10, 6))
plt.hist(df_spain['Kilometres'], bins=30, color='orange', edgecolor='black')
plt.title('Distribución del Kilometraje de Autos en España')
plt.xlabel('Kilometraje')
plt.ylabel('Frecuencia')
plt.grid()
plt.show()

#Este gráfico muestra la distribución del kilometraje de los autos en España.
```

El histograma revela una fuerte concentración de autos con bajo kilometraje, con la frecuencia máxima observada en los primeros rangos. La frecuencia disminuye drásticamente conforme se incrementa el kilometraje, lo que sugiere que hay menos autos disponibles en el mercado con alto kilometraje. Esto podría indicar que los compradores prefieren autos que han sido poco usados o que hay una oferta mayor de vehículos en buenas condiciones.

```python
# Gráfico de dispersión entre precio y año
plt.figure(figsize=(10, 6))
plt.scatter(df_spain['Year'], df_spain[' Price '], alpha=0.5, color='green')
plt.title('Relación entre Precio y Año de los Autos')
plt.xlabel('Año')
plt.ylabel('Precio')
plt.grid()
plt.show()

#Este gráfico ayuda a visualizar la relación entre el precio de los autos y su año de fabricación.
```

El gráfico de dispersión ilustra la relación entre el precio y el año de fabricación de los autos, mostrando una tendencia general en la que los autos más nuevos tienden a estar asociados con precios más altos. Sin embargo, hay una dispersión significativa en los precios, lo que sugiere que, aunque la edad del auto puede influir en el precio, otros factores como la marca, el modelo y el tipo de combustible también juegan un papel importante en la determinación del precio final.
